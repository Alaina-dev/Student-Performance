```python
#2.	Question: "Write the code to create a new column 'grade' where marks >= 85 are 'A', 70-84 are 'B', 60-69 are 'C', and below 60 are 'D'."

dataframe["grade"]=dataframe["mark"].apply(
    lambda x:"A" if x>=85 else
             "B" if x>=70 else
             "C" if x>=60 else
             "D"
)
dataframe
```
